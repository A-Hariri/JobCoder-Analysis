# Split the data into multiple parts

```python
import pandas as pd
import math

def split_csv(input_file, output_prefix, num_splits):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(input_file)

    # Calculate the number of lines in each split file
    num_lines = math.ceil(len(df) / num_splits)

    # Split the DataFrame into chunks
    split_dfs = [df[i:i+num_lines] for i in range(0, len(df), num_lines)]

    # Save each split DataFrame to a separate CSV file
    for i, split_df in enumerate(split_dfs):
        output_file = f"D:/pycharm projects/crawl linkedin/joindata/split/{output_prefix} cybersecurity2_{i + 1}.csv"
        split_df.to_csv(output_file, index=False)

if __name__ == "__main__":
    input_file = r"D:\pycharm projects\crawl linkedin\joindata\cybersecurity2.csv"
    
    # Replace with the name of your input CSV file
    output_prefix = "split_output_"  # Replace with the desired prefix for output files
    num_splits = 10

    split_csv(input_file, output_prefix, num_splits)
```
